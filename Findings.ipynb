```python
import pandas as pd
import ast
import networkx as nx
import matplotlib.pyplot as plt

# Read data from the Excel file
df = pd.read_excel("C:\\Users\\dorkf\\Desktop\\classified_users1.xlsx", sheet_name="Sheet1")

# Create an empty graph
G = nx.Graph()

# Add main nodes (users who tweeted)
for _, row in df.iterrows():
    G.add_node(row["name"], category=row["category"])

# Add secondary nodes (retweeters)
for _, row in df.iterrows():
    retweeters = ast.literal_eval(row["user_retweeters"])
    for retweeter in retweeters:
        G.add_node(retweeter)
        G.add_edge(row["name"], retweeter)

# Assign colors to nodes based on their categories
node_colors = []
node_sizes = []
for node, data in G.nodes(data=True):
    if "category" in data:
        if data["category"] == "General Influencers":
            node_sizes.append(100)
        elif data["category"] == "Stock-related Influencers":
            node_sizes.append(100)
        else:
            node_sizes.append(50)
        if data["category"] == "General Influencers":
            node_colors.append("yellow")
        elif data["category"] == "Stock-related Influencers":
            node_colors.append("red")
    else:
        node_colors.append("blue")
        node_sizes.append(50)

# Build the graph with the assigned colors and sizes
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=False, node_color=node_colors, node_size=node_sizes, font_size=8, font_weight="bold", alpha=0.8, edgecolors="black")
plt.show()
```

```python
import pandas as pd
import ast
import networkx as nx
import matplotlib.pyplot as plt

# Read data from the Excel file
df = pd.read_excel("C:\\Users\\dorkf\\Desktop\\classified_users3.xlsx", sheet_name="Sheet1")

# Create an empty graph
G = nx.Graph()

# Add main nodes (users who tweeted)
for _, row in df.iterrows():
    G.add_node(row["name"], category=row["category"])

# Add secondary nodes (retweeters)
for _, row in df.iterrows():
    retweeters = ast.literal_eval(row["user_retweeters"])
    for retweeter in retweeters:
        G.add_node(retweeter)
        G.add_edge(row["name"], retweeter)

# Calculate the Newman Modularity partition
newman_partition = nx.algorithms.community.label_propagation.label_propagation_communities(G)

# Create a dictionary to store the community of each user
user_community = {}
for idx, community in enumerate(newman_partition):
    for user in community:
        user_community[user] = idx

# Create a dictionary to store the count for each category
category_counts = {"General Influencers": 0, "Stock-related Influencers": 0}

# Calculate the count for each category
for node, data in G.nodes(data=True):
    if "category" in data:
        category = data["category"]
        if category in category_counts:
            category_counts[category] += 1

# Calculate the total number of users in the graph
total_users = len(G.nodes())

# Calculate the percentage of users in each category
category_percentages = {category: count / total_users * 100 for category, count in category_counts.items()}

# Calculate the percentage for "Amateurs"
amateurs_percentage = 100 - sum(category_percentages.values())

# Update the category_percentages dictionary with "Amateurs" percentage
category_percentages["Amateurs"] = amateurs_percentage

# Print the percentage of users in each category
for category, percentage in category_percentages.items():
    print(f"{category}: {percentage:.2f}%")
```
